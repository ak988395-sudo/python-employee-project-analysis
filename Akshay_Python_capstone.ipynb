```python
#Task1
import pandas as pd

# Creating project dataframe
project_data = {
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003",
           "A001","A003","A001","A004","A004","A005"],
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5",
                "Project 6","Project 7","Project 8","Project 9","Project 10",
                "Project 11","Project 12","Project 13","Project 14"],
    "Cost": [1002000,2000000,4500000,5500000,None,
             680000,400000,350000,None,300000,
             2000000,1000000,3000000,200000],
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished",
               "Failed","Finished","Failed","Ongoing","Finished",
               "Failed","Ongoing","Finished","Finished"]
}

project_df = pd.DataFrame(project_data)
project_df.to_csv("project_dataframe.csv", index=False)

# Creating employee dataframe
employee_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
    "Gender": ["M","F","M","F","F"],
    "City": ["Paris","London","Berlin","Newyork","Madrid"],
    "Age": [25,27,29,31,30]
}

employee_df = pd.DataFrame(employee_data)
employee_df.to_csv("employee_dataframe.csv", index=False)

# Creating seniority dataframe
seniority_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Designation Level": [2,2,3,2,3]
}

seniority_df = pd.DataFrame(seniority_data)
seniority_df.to_csv("seniority_level_dataframe.csv", index=False)

print(project_df)
print(employee_df)
print(seniority_df)
```

```python
#Task 2
# Convert Cost column into list so we can loop and update values
cost_list = project_df["Cost"].tolist()

# Variables to store running total and count
total = 0
count = 0

# Loop through each index of the cost list
for i in range(len(cost_list)):
    
    # If cost value exists (not NaN)
    if not pd.isna(cost_list[i]):
        total = total + cost_list[i]
        count = count + 1
    
    # If cost value is missing, replace with running average
    else:
        cost_list[i] = total / count

# Updating dataframe with new cost values
project_df["Cost"] = cost_list

# Saving updated dataframe back to CSV
project_df.to_csv("project_dataframe.csv", index=False)

print(project_df)
```

```python
#Task 3
# Lists to store first and last names separately
first_names = []
last_names = []

# Loop through each full name
for name in employee_df["Name"]:
    
    # Split name using space
    parts = name.split(" ")
    
    # Store first and last name
    first_names.append(parts[0])
    last_names.append(parts[1])

# Adding new columns to dataframe
employee_df["First Name"] = first_names
employee_df["Last Name"] = last_names

# Removing original Name column
employee_df.drop("Name", axis=1, inplace=True)

print(employee_df)
```

```python
#Task 4
# Merge project and employee data using ID
Final = pd.merge(project_df, employee_df, on="ID")

# Merge the result with seniority data
Final = pd.merge(Final, seniority_df, on="ID")

print(Final)
```

```python
#Task 5
# List to store bonus values
bonus = []

# Loop through each record
for i in range(len(Final)):
    
    # If project status is Finished, calculate 5% bonus
    if Final.loc[i, "Status"] == "Finished":
        bonus.append(Final.loc[i, "Cost"] * 0.05)
    else:
        bonus.append(0)

# Adding Bonus column to dataframe
Final["Bonus"] = bonus

print(Final)
```

```python
#Task 6
for i in range(len(Final)):
    
    # If project is Failed, reduce designation level by 1
    if Final.loc[i, "Status"] == "Failed":
        Final.loc[i, "Designation Level"] -= 1

# Removing rows where designation level is greater than 4
Final = Final[Final["Designation Level"] <= 4]

print(Final)
```

```python
#Task 7
# Loop through each employee
for i in range(len(Final)):
    
    # Add prefix based on gender
    if Final.loc[i, "Gender"] == "M":
        Final.loc[i, "First Name"] = "Mr. " + Final.loc[i, "First Name"]
    else:
        Final.loc[i, "First Name"] = "Mrs. " + Final.loc[i, "First Name"]

# Dropping Gender column
Final.drop("Gender", axis=1, inplace=True)

print(Final)
```

```python
#Task 8
for i in range(len(Final)):
    
    # If age is greater than 29, increase designation level
    if Final.loc[i, "Age"] > 29:
        Final.loc[i, "Designation Level"] += 1

print(Final)
```

```python
#Task 9
# Get unique employee IDs
ids = Final["ID"].unique()

# Lists to store final output
emp_ids = []
emp_names = []
total_costs = []

# Loop through each employee ID
for emp_id in ids:
    
    total = 0
    name = ""
    
    # Loop to sum project costs
    for i in range(len(Final)):
        if Final.loc[i, "ID"] == emp_id:
            total = total + Final.loc[i, "Cost"]
            name = Final.loc[i, "First Name"]
    
    # Store results
    emp_ids.append(emp_id)
    emp_names.append(name)
    total_costs.append(total)

# Creating final dataframe
TotalProjCost = pd.DataFrame({
    "ID": emp_ids,
    "First Name": emp_names,
    "Total Cost": total_costs
})

print(TotalProjCost)
```
